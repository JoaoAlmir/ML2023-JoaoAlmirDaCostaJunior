HOMICIDIO DOLOSO

LEITURA DE DADOS 

```python
import pandas as pd
df = pd.read_excel ('homicidios.xlsx')
df
```

REMOVENDO DADOS DUPLICADOS

```python
df = df.drop_duplicates()
df
```

FILTRANDO DADOS DE SP


```python
SP = df["MUNICIPIO_CIRCUNSCRICAO"] == "São Paulo"
df = df[SP]
df
```

TRANSFORMANDO STRINGS PARA LOWER CASE

```python
                          # se o tipo for string deixa minusculo
df = df.applymap(lambda s: s.lower() if type(s) == str else s)
df
```

PADRONIZANDO STRING

removendo acentuação e espaco

```python
import unidecode
                            #se o tipo for string remove os acentos
df = df.applymap(lambda s: unidecode.unidecode(s) if type(s) == str  else s)
                            #se o tipo for string remove os espacos
df = df.applymap(lambda s: s.replace(' ' , '') if type(s) == str  else s)
df
```

datetime64 para string
